```python
import numpy as np
import torch
from PIL import Image
import pandas as pd
import clip
import seaborn as sns

device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)
text_list = ['person', 'car', 'green traffic light', 'red traffic light', 'stop sign', 'bycicle']

def apply_clip(image_dir, text_list):
    image = preprocess(Image.open(image_dir)).unsqueeze(0).to(device)
    text = clip.tokenize(text_list).to(device)
    with torch.no_grad():
        image_features = model.encode_image(image)
        text_features = model.encode_text(text)
        logits_per_image, logits_per_text = model(image, text)
        probs = logits_per_image.softmax(dim=-1).cpu().numpy()

        print('Image directory: ' + str(image_dir))
        print("Label probs:", probs[0])
    return probs[0]

image_dir = "examples/ped.jpg"
pedestrian_softmax = apply_clip(image_dir, text_list)
```

```python
car_softmax = apply_clip("examples/car.jpg", text_list)
```

### Conformal Prediction on Perception Uncertainty

```python
df = pd.read_csv('distributions/nc_scores.csv')
ax = sns.histplot(data=df, x="scores", stat='proportion', bins=50, color='purple')
ax.set_xlim(-0.02, 1)
ax.set_xlabel('Softmax')
```

```python
def estimate_prob_guarantee(softmax_vector, data, col_name = 'scores', obj='car'):
    s = data[col_name]
    total = len(s)
    count = 0
    conf_score = 1 - np.sort(softmax_vector)[-2]
    for i in s:
        if i > 1 - conf_score:
            count += 1
    print('The perception uncertainty score (probability of correctly identifying '+obj +') is ' + str(count/total))
    return count/total

estimate_prob_guarantee(pedestrian_softmax, df, 'person')
estimate_prob_guarantee(car_softmax, df, 'scores')
```

### Conformal Prediction on Decision Uncertainty

```python
# We use the scenario from the Appendix as an example
'''
User: Turn right.
Assistant:
    A) wait for the car, then turn right
    B) wait for the car and pedestrian, then turn right
    C) turn right
    D) wait for the green traffic light, then turn right

User: Which option is correct? Answer with a single letter.
Assistant: A
'''
# Suppose the softmax scores for the last response is [A: 0.64, B: 0.18, C: 0.07, D: 0.11]
choice_vector = [0.64, 0.18, 0.07, 0.11]
```

```python
df1 = pd.read_csv('distributions/conf_pred.csv')
ax = sns.histplot(data=df1, x="Softmax", stat='proportion', bins=50)
ax.set_xlim(-0.02, 1)
```

```python
def estimate_decision_guarantee(softmax_vector, data, col_name = 'Softmax'):
    s = data[col_name]
    total = len(s)
    count = 0
    conf_score = 1 - np.sort(softmax_vector)[-2]
    for i in s:
        if i > 1 - conf_score:
            count += 1
    print('The decision uncertainty score (probability of the plan satisfying all specs) is ' + str(count/total))
    return count/total

estimate_prob_guarantee(choice_vector, df1, 'Softmax')
```
